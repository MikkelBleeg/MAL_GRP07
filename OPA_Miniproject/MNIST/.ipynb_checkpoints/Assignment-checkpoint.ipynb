# Loading Data

```python
import LoadData as loadD
import NearestCentroid as NearestCentroidMethod
import Visualization as VisualizationMethod
import KMeans as KmeansMethod
from sklearn.decomposition import PCA
import numpy as np
import KNeighbors as KNeighborsMethod
import BackPropagation as BackPropMethod
import MeanSquaredError as MeanSquaredErrorMethod
```

```python
Orl_data_train, Orl_data_test, Orl_label_train, Orl_label_test = loadD.LoadOrlData()
```

```python
MNIST_data_train, MNIST_label_train, MNIST_data_test, MNIST_label_test = loadD.LoadMNISTData()
```

```python
#Transforming MNIST datset to 2D PCA 
pca = PCA(n_components=2)
MNIST_data_train_pca = pca.fit_transform(MNIST_data_train)
MNIST_data_test_pca = pca.fit_transform(MNIST_data_test)

#Transforming Orldatset to 2D PCA 
pca = PCA(n_components=2)
Orl_data_train_pca = pca.fit_transform(Orl_data_train)
Orl_data_test_pca = pca.fit_transform(Orl_data_test)
```

```python
#Visualizeing the data 

#A visual representation of the first 6 samples for MNIST
VisualizationMethod.IllustrateMNIStData()

#MNIST 2D PCA
VisualizationMethod.VisualizePcaMNIST(MNIST_data_train_pca,MNIST_label_train)
#ORL 2D PCA
VisualizationMethod.VisualizePcaORL(Orl_data_train_pca,Orl_label_train)
```
